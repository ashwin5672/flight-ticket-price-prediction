# Statistical Test for Impact on Price by Numerical Columns
## Checking for Normality of Data.
* As we have a very large no. of records, we prefer to use Jarque-Bera Test for testing whether the data are drawn from normally distributed populations.
### Jarque Bera Test

* Null Hypothesis :  Data is Normally distributed.
* Alternate Hypothesis : Data is not Normally distributed.

```python
fp_num = fp.select_dtypes(include = np.number).columns
fp_num
```

```python
for i in fp_num:
    print( i, ':' ,stats.jarque_bera(fp[i].dropna()))
```

* As p_values for all features are less than significance levels of 0.05, we reject the null hypothesis that the data are drawn from normal populations.

* Now let us do both Pearson- R and Spearman Rank Correlation tests to validate the significance of the independent variables with the target feature -price. 

### Pearson - r Correlation Test.

This test calculates the pearson-r correlation coefficient between the given samples and then does a t-test to assume any statistcally significant correlation between the populations.

* Null Hypothesis : pearson-r coefficient = 0 and there is no correaltion between the populations of the 2 features.
* Alternate Hypothesis : pearson-r coefficient is <> 0 and there is a correaltion between the popilations of the 2 features.

```python
for i in fp_num:
    print(i, ':', stats.pearsonr(fp[i], fp['price']))
```

* As p-values  values of all features are less than 0.05, the null hypothesis is rejected and hence there is a statistically significant correlation between the predictors and target.

### Spearman Rank Correlation Test.

* This test takes the correlation of the ranks of the data points in each sample  thus makeing the test robust to outliers and reduces the sensitivity to the nature of distribution of data.

* Null Hypothesis : There is no correaltion between the populations of the 2 features.
* Alternate Hypothesis : There is a correaltion between the popilations of the 2 features.

```python
for i in fp_num:
    print(i, ':', stats.spearmanr(fp[i], fp['price']))
```

# Statsitical Tests for Impact on Price by Categorical Columns.

### Kruskal -Wallis Test

* This test is a non-parametric variant of f-oneway anova that does not require the assumptions of normality and heteroskedasticity of data.
* This test uses the rank sums as the central tendency which makes it robust to the outliers and nature of distribution of data.
* As our data is not normal, we proceed forward with this non- parametric test.

```python
airline = fp_cat_multi.groupby('airline')
stats.kruskal(airline.get_group('SpiceJet')['price'],
             airline.get_group('AirAsia')['price'],
             airline.get_group('Vistara')['price'],
             airline.get_group('GO_FIRST')['price'],
             airline.get_group('Indigo')['price'],
             airline.get_group('Air_India')['price'])
```

```python
s_city = fp_cat_multi.groupby('source_city')
stats.kruskal(s_city.get_group('Delhi')['price'],
             s_city.get_group('Mumbai')['price'],
             s_city.get_group('Bangalore')['price'],
             s_city.get_group('Kolkata')['price'],
             s_city.get_group('Hyderabad')['price'],
             s_city.get_group('Chennai')['price'])
```

```python
d_time = fp_cat_multi.groupby('departure_time')
stats.kruskal(d_time.get_group('Evening')['price'],
             d_time.get_group('Early_Morning')['price'],
             d_time.get_group('Morning')['price'],
             d_time.get_group('Afternoon')['price'],
             d_time.get_group('Night')['price'],
             d_time.get_group('Late_Night')['price'])
```

```python
a_time = fp_cat_multi.groupby('arrival_time')
stats.kruskal(a_time.get_group('Evening')['price'],
             a_time.get_group('Early_Morning')['price'],
             a_time.get_group('Morning')['price'],
             a_time.get_group('Afternoon')['price'],
             a_time.get_group('Night')['price'],
             a_time.get_group('Late_Night')['price'])
```

```python
d_city = fp_cat_multi.groupby('destination_city')
stats.kruskal(d_city.get_group('Delhi')['price'],
             d_city.get_group('Mumbai')['price'],
             d_city.get_group('Bangalore')['price'],
             d_city.get_group('Kolkata')['price'],
             d_city.get_group('Hyderabad')['price'],
             d_city.get_group('Chennai')['price'])
```

```python
stops = fp_cat_multi.groupby('stops')
stats.kruskal(stops.get_group('zero')['price'],
             stops.get_group('one')['price'],
             stops.get_group('two_or_more')['price'])
```

```python
new_c = fp_cat_multi.groupby('new_col')
stats.kruskal(new_c.get_group('less than 10 days')['price'],
             new_c.get_group('less than 20 days')['price'],
             new_c.get_group('less than 30 days')['price'],
             new_c.get_group('less than 40 days')['price'],
             new_c.get_group('less than 50 days')['price'])
```

* As p-value for all categories are less than the significant level of 0.05, we reject the null hypothesis and conclude that not all the sub-categories have the same rank sums.

* This shows that all the categories above has a statistically significant influence on the target - price.

###  Mann Whitney - U Test.

* As our target variable price is not drawn from a normal population, we use a non-parametric test call Mann-Whitney -U or Wilcoxon- Rank-Sum Test to determine the statistical signifiance of a binary categorical classification of a continuous numerical variable.

* Null Hypotheis - The rank sums for both classes are same. There is no impact of the category on numerical variable.
* Alternate Hypotheis - The rank sums for both classes are different. There is a significant impact of category on numerical variable.

```python
stats.mannwhitneyu(fp[fp['class'] == 'Economy']['price'],
                   fp[fp['class'] == 'Business']['price'])
```

* As the p-value is less than the significant levl of 0.05, we reject the null hypothesis that there is no dependence between class and price.
* The class category has a statistically significant impact on the price variable.
